```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

sp = SparkSession.builder.master("local[4]") \
            .appName("SparkApp-002") \
            .getOrCreate()

sc = sp.sparkContext
```

```python
ratingsFile = sc.textFile("ratings.csv")
header = ratingsFile.first()
ratingsRDD = ratingsFile.filter(lambda x: x != header).map(lambda l: l.split(","))

moviesFile = sc.textFile("movies.csv")
moviesHeader = moviesFile.first()
moviesRDD = moviesFile.filter(lambda x: x != moviesHeader).map(lambda l: l.split(","))

tagsFile = sc.textFile("tags.csv")
tagsHeader = tagsFile.first()
tagsRDD = tagsFile.filter(lambda x: x != tagsHeader).map(lambda l: l.split(","))


ratingsDF = sp.read.csv("ratings.csv", header=True, inferSchema=True)
```

```python
mappedRatingsRDD = ratingsRDD.map(lambda x: (int(x[1]), (int(x[0]), x[2])))
mappedTagsRDD = tagsRDD.map(lambda x: (int(x[1]), x[2]))
mappedMoviesRDD = moviesRDD.map(lambda x: (int(x[0]), x[1]))
mappedMoviesRDD.join(mappedTagsRDD).reduceByKey(lambda x, y: x + y).collect()
```
