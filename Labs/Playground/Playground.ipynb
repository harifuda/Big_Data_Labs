```python
ratingsFile = sc.textFile('ratings.csv')
ratingsHeader = ratingsFile.first()
ratingsRDD = ratingsFile.filter(lambda x: x != ratingsHeader).map(lambda x: (x.split(",")[1], x.split(",")[2]))
```

```python
ratingsRDD.collect()

moviesFile = sc.textFile('movies.csv')
moviesHeader = moviesFile.first()
moviesRDD = moviesFile.filter(lambda x: x != moviesHeader).map(lambda x: (int(x.split(",")[0]), x.split(",")[1]))

ratingsCountRDD = ratingsRDD.map(lambda x: ((int(x[0]), 1))) \
                            .reduceByKey(lambda x, y: x + y)

resultRDD = moviesRDD.join(ratingsCountRDD)
sorted(resultRDD.collect())
```
